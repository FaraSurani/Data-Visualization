```python
bike.weathersit.unique()    #we had four situations
```

```python
bike.weathersit.value_counts()  #this is called frequency table
```

```python
bike.weathersit.value_counts(normalize = True)*100
```

```python
(bike.weathersit.value_counts(normalize = True)*100).plot.pie()
#though it should be a bar chart
```

```python
(bike.weathersit.value_counts(normalize=True) * 100).plot.pie(autopct='%1.1f%%')
plt.title("Weather Situation")
plt.legend(["Clear Weather", "Cloudy Weather", "Light showers"])
plt.show()
```

```python
bike.weathersit.value_counts().plot.bar()
plt.xticks(rotation = 0)
plt.title("Weather Situation")
plt.xticks(ticks = [0,1,2,],labels = ["Clear Weather", "Cloudy Weather", "Light showers"])
plt.show
```
