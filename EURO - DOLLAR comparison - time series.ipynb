```python
euro  = pd.read_csv("euro.csv")
```

```python
euro_dollar = euro[[r'Period\Unit:', r'[US dollar ]']].copy()
```

```python
euro_dollar.rename(columns = {r'Period\Unit:':"Date", r'[US dollar ]':"USD"}, inplace = True)
```

```python
euro_dollar
```

```python
euro_dollar.Date = pd.to_datetime(euro_dollar.Date)
```

```python
euro_dollar
```

```python
euro_dollar.sort_values(by = "Date", inplace = True)
```

```python
euro_dollar
```

```python
euro_dollar.reset_index(drop = True, inplace = True)
```

```python
euro_dollar
```

```python
euro_dollar.USD = euro_dollar.USD.astype(float)
```

```python
rolling = [7,30,90,180,365]
plots = [2,3,4,5,6]
```

```python
plt.figure(figsize = (11,9))
plt.subplot(3,2,1)
plt.plot(euro_dollar["Date"], euro_dollar.USD)
plt.title("Original")

for rol, plot in zip(rolling, plots):
    plt.subplot(3,2,plot)
    plt.plot(euro_dollar["Date"], euro_dollar.USD.rolling(rol).mean(), linewidth = 2)
    plt.title(f"Moving Average {rol}")
    
plt.tight_layout()
plt.show()
```

```python
euro_dollar["Rolling_mean_30"] = euro_dollar.USD.rolling(30).mean()
```

```python
euro_dollar
```

```python
bol_2006_2009 = (euro_dollar.Date.dt.year>= 2006) & (euro_dollar.Date.dt.year <= 2009)
```

```python
bol_2007_2008 = (euro_dollar.Date.dt.year>= 2007) & (euro_dollar.Date.dt.year <= 2008)
```

```python
financial_crisis = euro_dollar[bol_2006_2009].copy()
```

```python
financial_crisis
```

```python
financial_crisis_07_08 = financial_crisis[bol_2007_2008].copy()
```

```python
fix, ax = plt.subplots(figsize = (8,3))

ax.plot(financial_crisis["Date"], financial_crisis.Rolling_mean_30, linewidth = 1)
ax.plot(financial_crisis_07_08["Date"], financial_crisis_07_08.Rolling_mean_30, linewidth = 3,
       color = "#b00b1e")

plt.show()
```
