```python
weekend = day[~ (day["dayofweek"] <= 4)]
weekend                # telda ~ is sign of not equal
```

```python
weekdays.pivot_table(index = "hour", values="traffic_volume",)
```

```python
plt.figure(figsize = (11,4))

#weekday graph
plt.subplot(1,2,1)

hourly_weekday_avg = weekdays.pivot_table(index = "hour", values="traffic_volume")
plt.plot(hourly_weekday_avg.index, hourly_weekday_avg)
plt.ylim(1500,6500)

#weekend graph

plt.subplot(1,2,2)

hourly_weekend_avg = weekend.pivot_table(index = "hour", values="traffic_volume")
plt.plot(hourly_weekend_avg.index, hourly_weekend_avg)
plt.ylim(1500,6500)  #setting up same scale on both graphs to have comparision

plt.show()
```
